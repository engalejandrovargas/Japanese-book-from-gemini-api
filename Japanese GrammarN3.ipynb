```python
words.dropna(axis=0, inplace=True)
```

```python
words_list = words['文法レッスン'].tolist()
```

```python
import google.generativeai as genai

genai.configure(api_key="YOUR-APIKEY")

# Set up the model
generation_config = {
  "temperature": 0.5,
  "top_p": 1,
  "top_k": 1
}

model = genai.GenerativeModel(model_name="gemini-pro",
                              generation_config=generation_config)
def prompt_words(word):
    prompt_parts = [
  "input: Word in Japanese",
  "output: -Explain me the grammar easy, friendly and detailed, \n-use cases, \n-to whom can Use that word, \n-examples  \n-sinonims \n-antonims \nEverythin with hiragana and translation in English.",
  "input: ",
  "output: **Grammar:**\n\n書かれている (かかれている) is the passive form of the verb 書く (かく), which means \"to write.\" The passive form is used to indicate that the subject of the sentence is being acted upon by the verb. In this case, the subject is the thing that is being written on, and the verb is \"to be written.\"\n\n**Usage:**\n\n書かれている can be used in a variety of situations, such as:\n\n* To describe something that is written on a piece of paper, a wall, or another surface.\n* To describe something that is written in a book, a magazine, or a newspaper.\n* To describe something that is written in a computer file.\n\n**Who can use it:**\n\n書かれている can be used by anyone who wants to describe something that is written. It is a common verb that is used in everyday conversation.\n\nExamples**:\n\n* この手紙はきれいに書かれています。(This letter is neatly written.)\n* この文章はわかりやすく書かれています。(This article is easy to understand.)\n* この看板は日本語で書かれています。(This sign is written in Japanese.)\n\n**Synonyms**:\n\n* 書き記されている (かきとどめられている) - written down \n* 書きつけられている (かきつけられている) - written on\n* 筆記されている (ひっきされている) - written down \n\n**Antonyms**:\n\n* 書かれていない (かかれていない) - Not written down\n* 書き残されていない (かきのこされていない) - Not written \n* 書き留められていない (かきとどめられていない) - Not written down",
  "input: 上げる",
  "output: **Grammar:**\n\n上げる (あげる) is a verb that means \"to raise\" or \"to lift.\" It can also be used to mean \"to give\" or \"to present.\" The verb is conjugated as follows:\n\n* Present tense: 上げる (あげる)\n* Past tense: 上げた (あげた)\n* Future tense: 上げる (あげる)\n\n**Usage:**\n\n上げる can be used in a variety of situations, such as:\n\n* To raise or lift an object.\n* To give or present something to someone.\n* To offer or propose something.\n* To increase or elevate something.\n\n**Who can use it:**\n\n上げる can be used by anyone who wants to express the idea of raising, lifting, giving, or presenting something. It is a common verb that is used in everyday conversation.\n\n**Examples**:\n\n* 彼は本をテーブルの上に上げた。(He raised the book onto the table.)\n* 私は彼女にプレゼントを上げた。(I gave her a present.)\n* 彼は私に結婚を申し上げた。(He proposed to me.)\n* 景気回復を目標に、政府は消費税率を上げた。(With the goal of economic recovery, the government raised the consumption tax rate.)\n\n**Synonyms**:\n\n* 掲げる (かかげる) - to raise\n* 提示する (ていしする) - to present\n* 提出する (ていしゅつする) - to submit\n* 献上する (けんじょうする) - to offer\n\n**Antonyms**:\n\n* 下げる (さげる) - lower\n* 降ろす (おろす) - take it down\n* 奪う (うばう) - take away\n* 隠す (かくす) - hide",
  "input:"+word,
  "output: ",
]
    response = model.generate_content(prompt_parts)
#     print(response.text)
#     print("\n\n-------------------------------------------\n\n")
    return response.text + "\n\n----\n\n"
```

```python
len(words_list)
```

```python
my_book = {}

for i in range(len(words_list)):
    my_book[str(i)] = i
    my_book[words_list[i]] = prompt_words(words_list[i])
```

```python
len(my_book)
```

```python
my_book
```

```python
start_index = 10
end_index = 14

# Get the keys within the specified range
keys_to_iterate = list(my_book.keys())[start_index:end_index]

# Iterate over the specified range of keys
for key in keys_to_iterate:
    print(f'Key: {key}, Value: {my_book[key]}')
```

```python
filtered_dict = {key: value for key, value in my_book.items() if not key.isdigit()}
```

```python
from docx import Document

# Create a new Word document
doc = Document()

# Iterate over the dictionary and add entries to the document
for key, value in filtered_dict.items():
    # Add the key as a heading
    doc.add_heading(f'{key}', level=1)

    # Add the value (assuming it is a string) as a paragraph
    if isinstance(value, str):
        doc.add_paragraph(value)

# Save the document to a Word file
doc.save('document.docx')
```
